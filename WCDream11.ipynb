```python
def sel_players(Ground,team):
    a=df[df["Ground"].isin(Ground)]
    b=a[a["Batting_Team"].isin(team)]
    b["Runs"]=b["Runs"].astype(int)
    trp=b.groupby(["batsmen","Batting_Team","Ground"])["Runs"].sum().to_frame().reset_index().rename(columns={"Runs":"Total Runs"})
    ap=b.groupby(["batsmen","Batting_Team","Ground"])["Runs"].mean().to_frame().reset_index().rename(columns={"Runs":"Average"})
    mp=b.groupby(["batsmen","Batting_Team","Ground"])["Runs"].median().to_frame().reset_index().rename(columns={"Runs":"Median"})
    minip=b.groupby(["batsmen","Batting_Team","Ground"])["Runs"].min().to_frame().reset_index().rename(columns={"Runs":"Min Runs"})
    maxp=b.groupby(["batsmen","Batting_Team","Ground"])["Runs"].max().to_frame().reset_index().rename(columns={"Runs":"Max Runs"})
    stdp=b.groupby(["batsmen","Batting_Team","Ground"])["Runs"].std().to_frame().reset_index().rename(columns={"Runs":"std Deviation"})
    nmp=b["batsmen"].value_counts().to_frame().reset_index()
    nmp.rename(columns={"index":"batsmen","batsmen":"Matches Played"},inplace=True)
    f=trp.merge(ap,on=["batsmen","Batting_Team","Ground"],how="outer")
    f=f.merge(mp,on=["batsmen","Batting_Team","Ground"],how="outer")
    f=f.merge(minip,on=["batsmen","Batting_Team","Ground"],how="outer")
    f=f.merge(maxp,on=["batsmen","Batting_Team","Ground"],how="outer")
    f=f.merge(stdp,on=["batsmen","Batting_Team","Ground"],how="outer")
    f=f.merge(nmp,on=["batsmen"],how="outer")
    f=f.fillna(0)
    f["Co. eff. variance"]=f["std Deviation"]/f["Average"]
    f=f.fillna(0)
    return f 
```

```python
def sel_bowlers(Ground,team):
    a=bowler_stat[bowler_stat["Ground"].isin(Ground)]
    b=a[a["Bowling_Team"].isin(team)]
    #b["Runs"]=b["Runs"].astype(int)
    runs_leaked=b.groupby(["Bowler","Bowling_Team","Ground"])["Runs"].sum().to_frame().reset_index().rename(columns={"Runs":"Runs Leaked"})
    avgruns=b.groupby(["Bowler","Bowling_Team","Ground"])["Runs"].mean().to_frame().reset_index().rename(columns={"Runs":"Average Runs"})
    trp=b.groupby(["Bowler","Bowling_Team","Ground"])["Wicket"].sum().to_frame().reset_index().rename(columns={"Wicket":"Total Wickets"})
    ap=b.groupby(["Bowler","Bowling_Team","Ground"])["Wicket"].mean().to_frame().reset_index().rename(columns={"Wicket":"Average Wickets"})
    mp=b.groupby(["Bowler","Bowling_Team","Ground"])["Wicket"].median().to_frame().reset_index().rename(columns={"Wicket":"Median Wickets"})
    minip=b.groupby(["Bowler","Bowling_Team","Ground"])["Wicket"].min().to_frame().reset_index().rename(columns={"Wicket":"Min Wickets"})
    maxp=b.groupby(["Bowler","Bowling_Team","Ground"])["Wicket"].max().to_frame().reset_index().rename(columns={"Wicket":"Max Wickets"})
    stdp=b.groupby(["Bowler","Bowling_Team","Ground"])["Wicket"].std().to_frame().reset_index().rename(columns={"Wicket":"std Deviation"})
    nmp=b["Bowler"].value_counts().to_frame().reset_index()
    nmp.rename(columns={"index":"Bowler","Bowler":"Matches Played"},inplace=True)
    f=nmp.merge(runs_leaked,on=["Bowler"],how="outer")
    f=f.merge(avgruns,on=["Bowler","Bowling_Team","Ground"],how="outer")
    f=f.merge(trp,on=["Bowler","Bowling_Team","Ground"],how="outer")
    f=f.merge(ap,on=["Bowler","Bowling_Team","Ground"],how="outer")
    f=f.merge(mp,on=["Bowler","Bowling_Team","Ground"],how="outer")
    f=f.merge(minip,on=["Bowler","Bowling_Team","Ground"],how="outer")
    f=f.merge(maxp,on=["Bowler","Bowling_Team","Ground"],how="outer")
    f=f.merge(stdp,on=["Bowler","Bowling_Team","Ground"],how="outer")
    
    f=f.fillna(0)
    f["Co. eff. variance"]=f["std Deviation"]/f["Average Wickets"]
    f=f.fillna(0)
    f=f[["Bowler","Bowling_Team","Ground","Matches Played","Runs Leaked","Average Runs","Total Wickets","Average Wickets","Median Wickets","Min Wickets","Max Wickets","std Deviation","Co. eff. variance"]]
    return f 
```
